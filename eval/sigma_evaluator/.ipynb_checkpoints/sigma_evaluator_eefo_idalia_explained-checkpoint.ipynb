```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import torch
import earthkit.data as ekd

import sys
sys.path.append('/home/ecm5702/dev/downscaling-tools')

from get_objects_from_ckpt import (
    ObjectFromCheckpointLoader,
    get_checkpoint,
    instantiate_config,
    adapt_config_hpc,
)
from eval_suite.sigma_evaluator.sigma_evaluator import SigmaEvaluator
```

```python
# Paths / run config
DATA_TC = '/home/ecm5702/hpcperm/data/tc/idalia'
EXTRACTED_EEFO = '/home/ecm5702/hpcperm/data/tc/idalia/extracted_eefo_o96'
DIR_EXP = '/home/ecm5702/scratch/aifs/checkpoint'
NAME_EXP = "0e1602b36722471f97a3dfa93b81297e"
NAME_CKPT = 'anemoi-by_time-epoch_021-step_100000.ckpt'
DEVICE = 'cuda'

SIGMAS = [0.02, 0.1, 0.5, 1.0, 2.0, 100, 1000, 10000, 100000, 1000000]
N_BATCHES_MODEL = 1   # quick GPU run
SEED = 20260213

print('Checkpoint:', os.path.join(DIR_EXP, NAME_EXP, NAME_CKPT))
print('Extracted Idalia EEFO dir:', EXTRACTED_EEFO)
assert torch.cuda.is_available(), 'CUDA is required. Run this notebook on a GPU node/kernel.'
torch.backends.cudnn.benchmark = True
print('CUDA available:', torch.cuda.is_available())
print('GPU count:', torch.cuda.device_count())
print('Current GPU:', torch.cuda.get_device_name(0))
```

```python
# Load model + datamodule explicitly (no subprocess)
checkpoint, cfg_ckpt = get_checkpoint(DIR_EXP, NAME_EXP, NAME_CKPT)
cfg_base = instantiate_config()
cfg_ckpt = adapt_config_hpc(cfg_ckpt, cfg_base)

loader = ObjectFromCheckpointLoader(DIR_EXP, NAME_EXP, NAME_CKPT)
loader.config_checkpoint = cfg_ckpt
loader.config_for_datamodule.dataloader.validation.frequency = '50h'
loader.load()

datamodule = loader.datamodule
downscaler = loader.downscaler.to(DEVICE)
interface = loader.interface.to(DEVICE)

sigma_eval = SigmaEvaluator(downscaler, datamodule, N_samples=N_BATCHES_MODEL)

dates = datamodule.ds_valid.data.dates
print('Validation date axis example:')
print(' first:', dates[0])
print(' last :', dates[-1])
print(' len  :', len(dates))
```

```python
# Sample from the model with explicit sampler params and inspect outputs
dl = datamodule.val_dataloader()
batch = next(iter(dl))
x_in, x_in_hres, y = [t.to(DEVICE) for t in batch]

noise_scheduler_params = {
    'schedule_type': 'karras',
    'sigma_max': 80.0,
    'sigma_min': 0.03,
    'rho': 7.0,
    'num_steps': 10,
}
sampler_params = {
    'sampler': 'heun',
    'S_churn': 2.5,
    'S_min': 0.75,
    'S_max': 80.0,
    'S_noise': 1.05,
}

print('x_in shape    :', tuple(x_in.shape))
print('x_in_hres shape:', tuple(x_in_hres.shape))
print('y shape       :', tuple(y.shape))
print('noise_scheduler_params:', noise_scheduler_params)
print('sampler_params:', sampler_params)

n_members = 2
samples = []
for m in range(n_members):
    print(f'Generating sample member {m+1}/{n_members} ...')
    with torch.inference_mode(), torch.autocast(device_type='cuda'):
        pred = interface.predict_step(
            x_in[0:1].clone(),
            x_in_hres[0:1].clone(),
            noise_scheduler_params=noise_scheduler_params,
            sampler_params=sampler_params,
        )
    samples.append(pred.detach().cpu())

print('Done. Sample tensor shapes:', [tuple(s.shape) for s in samples])
```

```python
# Basic TC proxy metrics from sampled outputs (if fields exist)
name_to_idx = datamodule.data_indices.model.output.name_to_index
print('Has variables:', list(name_to_idx.keys())[:20], '...')

i_u = name_to_idx.get('10u', None)
i_v = name_to_idx.get('10v', None)
i_m = name_to_idx.get('msl', None)

proxy_rows = []
for k, s in enumerate(samples, start=1):
    # expected shape ~ [1, 1, grid, var] or compatible
    arr = s.numpy()
    arr = np.squeeze(arr)
    if arr.ndim == 2:
        # [grid, var]
        pass
    elif arr.ndim > 2:
        # fallback: take first axes until [grid, var]
        arr = arr.reshape(-1, arr.shape[-1])

    row = {'member': k}
    if i_m is not None and i_m < arr.shape[1]:
        row['min_msl_proxy'] = float(np.nanmin(arr[:, i_m]))
    if i_u is not None and i_v is not None and i_u < arr.shape[1] and i_v < arr.shape[1]:
        wspd = np.sqrt(arr[:, i_u] ** 2 + arr[:, i_v] ** 2)
        row['max_wind_proxy'] = float(np.nanmax(wspd))
    proxy_rows.append(row)

display(pd.DataFrame(proxy_rows))
```

```python
# Intensity proxy on model samples vs target (high-TC input context)
# Assumes sampling cell already produced: samples, y, datamodule

import numpy as np
import pandas as pd

name_to_idx = datamodule.data_indices.model.output.name_to_index
i_u = name_to_idx.get("10u")
i_v = name_to_idx.get("10v")
i_m = name_to_idx.get("msl")

if None in (i_u, i_v, i_m):
    raise KeyError("Need output vars: 10u, 10v, msl")

def to_grid_var(a):
    a = np.asarray(a)
    a = np.squeeze(a)
    if a.ndim == 2:
        return a
    if a.ndim > 2:
        return a.reshape(-1, a.shape[-1])
    raise ValueError(f"Unexpected shape: {a.shape}")

def tc_metrics(arr):
    msl = arr[:, i_m]
    wspd = np.sqrt(arr[:, i_u] ** 2 + arr[:, i_v] ** 2)
    min_msl = float(np.nanmin(msl))
    max_wind = float(np.nanmax(wspd))
    intensity = ((1010.0 - min_msl) / 60.0 + (max_wind / 50.0)) / 2.0
    return min_msl, max_wind, float(intensity)

rows = []

# Target baseline from current batch (sample 0)
arr_y = to_grid_var(y[0].detach().cpu().numpy())
y_msl, y_wind, y_int = tc_metrics(arr_y)
rows.append({
    "source": "target_y",
    "member": 0,
    "min_msl_hPa": y_msl,
    "max_wind_ms": y_wind,
    "intensity_score": y_int,
})

# Model samples
for k, s in enumerate(samples, start=1):
    arr = to_grid_var(s.detach().cpu().numpy())
    msl, wind, inten = tc_metrics(arr)
    rows.append({
        "source": "sample",
        "member": k,
        "min_msl_hPa": msl,
        "max_wind_ms": wind,
        "intensity_score": inten,
        "d_msl_vs_target": msl - y_msl,
        "d_wind_vs_target": wind - y_wind,
        "d_int_vs_target": inten - y_int,
    })

df = pd.DataFrame(rows)
display(df)

samples_df = df[df["source"] == "sample"]
print("\nSample intensity summary:")
print(samples_df[["min_msl_hPa", "max_wind_ms", "intensity_score"]].describe())
```

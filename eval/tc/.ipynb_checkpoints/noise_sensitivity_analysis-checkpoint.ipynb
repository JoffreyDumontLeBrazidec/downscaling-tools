# TC Noise Sensitivity Analysis

This notebook analyzes the sensitivity of tropical cyclone fields to Gaussian noise perturbations.

## Objectives:
1. Load the strongest TC instance (highest wind speed)
2. Add Gaussian noise with varying sigma (std) and mean=0
3. Visualize original vs noised fields at different noise levels
4. Check statistical characteristics: does it look like pure Gaussian noise or still retain cyclone structure?
5. Determine the sigma threshold where cyclone structure becomes indistinguishable from noise

```python
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from scipy import stats
import earthkit.data as ekd

# Add the current directory to path
sys.path.insert(0, os.getcwd())

from identify_strong_events import StrongEventIdentifier, CycloneInstance

%load_ext autoreload
%autoreload 2
```

## Configuration

```python
# Data configuration
DATA_DIR = "/home/ecm5702/hpcperm/data/tc"
EVENT_NAME = "idalia"
EXPID = "ENFO_O320_0001"  # or "EEFO_O96_0001"

# Noise levels to test (standard deviations in m/s for wind field)
# Start with small noise and gradually increase
SIGMA_LEVELS = [0.5, 1.0, 2.0, 5.0, 10.0, 15.0, 20.0]
```

## Step 1: Load Strongest TC Instance (Highest Wind Speed)

```python
# Load strongest instances from the analysis
# You should have run the identify_strongest_instances first
identifier = StrongEventIdentifier(DATA_DIR, EVENT_NAME)

DATES = ["20230826", "20230827", "20230828", "20230829", "20230830"]
N_MEMBERS = 10

# Identify strongest instances
print("Identifying strongest TC instances...")
strongest_instances = identifier.identify_strongest_instances(EXPID, DATES, N_MEMBERS, top_k=50)

# Sort by max wind speed to get the instance with highest wind
strongest_by_wind = sorted(strongest_instances, key=lambda x: x.max_wind, reverse=True)

# Get the strongest one
strongest = strongest_by_wind[0]
print(f"\n{'='*80}")
print(f"STRONGEST TC BY WIND SPEED:")
print(f"  Date: {strongest.base_time}")
print(f"  Forecast Step: +{strongest.forecast_step}h")
print(f"  Member: {strongest.member}")
print(f"  Max Wind: {strongest.max_wind:.2f} m/s")
print(f"  Min MSLP: {strongest.min_mslp:.2f} hPa")
print(f"  Location: ({strongest.center_lat:.2f}°N, {strongest.center_lon:.2f}°E)")
print(f"  Intensity Score: {strongest.intensity_score:.3f}")
print(f"{'='*80}\n")
```

```python
# Load the actual GRIB data for this specific instance
stream = EXPID.split("_")[0]
grid = EXPID.split("_")[1]
version = EXPID.split("_")[-1]
filename = f"surface_pf_{stream}_{grid}_{version}_{strongest.base_time}.grib"
filepath = os.path.join(DATA_DIR, EVENT_NAME, filename)

print(f"Loading data from: {filepath}")
data = ekd.from_source("file", filepath).to_xarray(engine="cfgrib")

# Extract the specific forecast step and member
isel_dict = {}
if "number" in data.dims:
    isel_dict["number"] = strongest.member
if "time" in data.dims:
    isel_dict["time"] = 0
if "surface" in data.dims:
    isel_dict["surface"] = 0

# Get step index
steps = data["step"].values
step_idx = np.where(steps == np.timedelta64(strongest.forecast_step, "h"))[0][0]
isel_dict["step"] = step_idx

# Extract wind components
u10 = np.squeeze(data["u10"].isel(**isel_dict).values)
v10 = np.squeeze(data["v10"].isel(**isel_dict).values)
wind_field_original = np.sqrt(u10**2 + v10**2)

# Extract MSLP for reference
mslp_original = np.squeeze(data["msl"].isel(**isel_dict).values)
if np.nanmean(mslp_original) > 50000:
    mslp_original = mslp_original / 100  # Convert Pa to hPa

# Get coordinates
lats = data["latitude"].values
lons = data["longitude"].values

print(f"✓ Data loaded successfully")
print(f"  Wind field shape: {wind_field_original.shape}")
print(f"  Wind range: [{np.nanmin(wind_field_original):.2f}, {np.nanmax(wind_field_original):.2f}] m/s")
print(f"  Wind mean: {np.nanmean(wind_field_original):.2f} m/s")
print(f"  Wind std: {np.nanstd(wind_field_original):.2f} m/s")
```

## Step 2: Add Gaussian Noise with Varying Sigma

```python
def add_gaussian_noise(field, sigma, seed=42):
    """
    Add Gaussian noise with mean=0 and std=sigma to a field

    Args:
        field: Original field (numpy array)
        sigma: Standard deviation of Gaussian noise
        seed: Random seed for reproducibility

    Returns:
        Noised field
    """
    np.random.seed(seed)
    noise = np.random.normal(loc=0.0, scale=sigma, size=field.shape)
    return field + noise, noise


def analyze_noise_characteristics(original, noised, noise, sigma):
    """
    Analyze statistical characteristics of noised field vs pure Gaussian noise

    Returns:
        Dictionary with statistics
    """
    stats_dict = {
        "sigma_target": sigma,
        "original_mean": np.nanmean(original),
        "original_std": np.nanstd(original),
        "original_max": np.nanmax(original),
        "noised_mean": np.nanmean(noised),
        "noised_std": np.nanstd(noised),
        "noised_max": np.nanmax(noised),
        "noise_mean": np.nanmean(noise),
        "noise_std": np.nanstd(noise),
        "signal_to_noise_ratio": np.nanstd(original) / sigma,
    }

    # Test if noised field follows Gaussian distribution (Shapiro-Wilk test on sample)
    # Sample 5000 points for the test (full field too large)
    sample_size = min(5000, noised.ravel().shape[0])
    sample_indices = np.random.choice(noised.ravel().shape[0], size=sample_size, replace=False)
    sample = noised.ravel()[sample_indices]
    sample = sample[~np.isnan(sample)]

    if len(sample) > 3:
        shapiro_stat, shapiro_p = stats.shapiro(sample)
        stats_dict["shapiro_p_value"] = shapiro_p
        stats_dict["looks_gaussian"] = shapiro_p > 0.05  # If p > 0.05, cannot reject Gaussian hypothesis

    return stats_dict


# Generate noised fields for all sigma levels
noised_fields = {}
noise_only = {}
statistics = []

print("Generating noised fields...")
for sigma in SIGMA_LEVELS:
    noised, noise = add_gaussian_noise(wind_field_original, sigma)
    noised_fields[sigma] = noised
    noise_only[sigma] = noise

    # Analyze characteristics
    stats_dict = analyze_noise_characteristics(wind_field_original, noised, noise, sigma)
    statistics.append(stats_dict)

    print(
        f"  σ={sigma:5.1f} m/s: SNR={stats_dict['signal_to_noise_ratio']:.3f}, "
        f"Max={stats_dict['noised_max']:.1f} m/s, "
        f"Gaussian-like: {stats_dict.get('looks_gaussian', 'N/A')}"
    )

# Convert to DataFrame for easy analysis
stats_df = pd.DataFrame(statistics)
print("\n✓ Noise generation complete")
```

## Step 3: Visualize Original vs Noised Fields

```python
# Create spatial plots comparing original vs noised fields
n_levels = len(SIGMA_LEVELS)
fig, axes = plt.subplots(2, 4, figsize=(20, 10), subplot_kw={"projection": ccrs.PlateCarree()})
axes = axes.flatten()

# Plot original in first panel
ax = axes[0]
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linestyle=":")
ax.set_title(f"Original Wind Field\nMax: {np.nanmax(wind_field_original):.1f} m/s", fontsize=12, fontweight="bold")

lats_flat = lats.ravel()
lons_flat = lons.ravel()
wind_flat = wind_field_original.ravel()

scatter = ax.tricontourf(lons_flat, lons_flat, wind_flat, levels=15, cmap="YlOrRd", transform=ccrs.PlateCarree())
ax.plot(
    strongest.center_lon, strongest.center_lat, "k*", markersize=15, transform=ccrs.PlateCarree(), label="TC Center"
)
ax.legend()
plt.colorbar(scatter, ax=ax, label="Wind Speed (m/s)", shrink=0.8)

# Plot noised fields
for i, sigma in enumerate(SIGMA_LEVELS):
    ax = axes[i + 1]
    ax.coastlines()
    ax.add_feature(cfeature.BORDERS, linestyle=":")

    noised = noised_fields[sigma]
    snr = stats_df.iloc[i]["signal_to_noise_ratio"]
    looks_gaussian = stats_df.iloc[i].get("looks_gaussian", "N/A")

    ax.set_title(f"σ={sigma:.1f} m/s (SNR={snr:.2f})\nGaussian-like: {looks_gaussian}", fontsize=10)

    noised_flat = noised.ravel()
    scatter = ax.tricontourf(lons_flat, lats_flat, noised_flat, levels=15, cmap="YlOrRd", transform=ccrs.PlateCarree())
    ax.plot(strongest.center_lon, strongest.center_lat, "k*", markersize=12, transform=ccrs.PlateCarree())
    plt.colorbar(scatter, ax=ax, label="Wind Speed (m/s)", shrink=0.8)

plt.tight_layout()
plt.suptitle("Wind Field Degradation with Increasing Gaussian Noise", y=1.00, fontsize=14, fontweight="bold")
plt.show()
```

## Step 4: Statistical Analysis - Distribution Comparison

```python
# Plot histograms to visualize distribution characteristics
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()

# Original field
ax = axes[0]
wind_flat_clean = wind_field_original.ravel()
wind_flat_clean = wind_flat_clean[~np.isnan(wind_flat_clean)]
ax.hist(wind_flat_clean, bins=50, density=True, alpha=0.7, color="blue", edgecolor="black")
ax.axvline(np.mean(wind_flat_clean), color="red", linestyle="--", label=f"Mean={np.mean(wind_flat_clean):.1f}")
ax.set_title("Original Wind Field\nDistribution", fontweight="bold")
ax.set_xlabel("Wind Speed (m/s)")
ax.set_ylabel("Probability Density")
ax.legend()
ax.grid(True, alpha=0.3)

# Noised fields with Gaussian overlay
for i, sigma in enumerate(SIGMA_LEVELS):
    ax = axes[i + 1]

    noised = noised_fields[sigma]
    noised_flat = noised.ravel()
    noised_flat = noised_flat[~np.isnan(noised_flat)]

    # Plot histogram
    ax.hist(noised_flat, bins=50, density=True, alpha=0.7, color="orange", edgecolor="black", label="Noised Field")

    # Overlay theoretical Gaussian with same mean and std
    mean_noised = np.mean(noised_flat)
    std_noised = np.std(noised_flat)
    x_range = np.linspace(mean_noised - 4 * std_noised, mean_noised + 4 * std_noised, 100)
    gaussian_pdf = stats.norm.pdf(x_range, mean_noised, std_noised)
    ax.plot(x_range, gaussian_pdf, "r-", linewidth=2, label="Perfect Gaussian")

    snr = stats_df.iloc[i]["signal_to_noise_ratio"]
    p_val = stats_df.iloc[i].get("shapiro_p_value", 0)

    ax.set_title(f"σ={sigma:.1f} m/s\nSNR={snr:.2f}, p={p_val:.4f}", fontsize=10)
    ax.set_xlabel("Wind Speed (m/s)")
    ax.set_ylabel("Probability Density")
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.suptitle("Distribution Analysis: Does the Noised Field Look Gaussian?", y=1.00, fontsize=14, fontweight="bold")
plt.show()

print("\nInterpretation:")
print("- If histogram matches red curve (Perfect Gaussian), field looks like pure noise")
print("- If histogram differs from red curve, field retains structure from original cyclone")
print("- Shapiro-Wilk test p-value > 0.05 suggests Gaussian distribution")
```

## Step 5: Summary Statistics and Threshold Identification

```python
# Display comprehensive statistics table
print("=" * 100)
print("NOISE SENSITIVITY ANALYSIS - SUMMARY")
print("=" * 100)
print(f"\nOriginal Wind Field Stats:")
print(f"  Mean: {np.nanmean(wind_field_original):.2f} m/s")
print(f"  Std:  {np.nanstd(wind_field_original):.2f} m/s")
print(f"  Max:  {np.nanmax(wind_field_original):.2f} m/s")
print(f"  Min:  {np.nanmin(wind_field_original):.2f} m/s")
print(f"\n{'-'*100}")

# Format and display the statistics DataFrame
display_df = stats_df[
    ["sigma_target", "signal_to_noise_ratio", "noised_max", "noised_std", "shapiro_p_value", "looks_gaussian"]
].copy()
display_df.columns = ["Sigma (m/s)", "SNR", "Max Wind (m/s)", "Std (m/s)", "Shapiro p-value", "Looks Gaussian"]
display_df = display_df.round(3)

print("\nNoise Level Statistics:")
print(display_df.to_string(index=False))
print(f"\n{'-'*100}")

# Identify threshold sigma where field becomes Gaussian-like
threshold_sigma = None
for i, row in stats_df.iterrows():
    if row.get("looks_gaussian", False):
        threshold_sigma = row["sigma_target"]
        break

if threshold_sigma:
    print(f"\n🎯 THRESHOLD IDENTIFIED:")
    print(
        f"   At σ = {threshold_sigma} m/s, the field becomes statistically indistinguishable from pure Gaussian noise"
    )
    print(
        f"   Signal-to-Noise Ratio at threshold: {stats_df[stats_df['sigma_target']==threshold_sigma]['signal_to_noise_ratio'].values[0]:.3f}"
    )
else:
    print(
        f"\n⚠️  No threshold found in tested range. Field retains structure even at highest σ = {SIGMA_LEVELS[-1]} m/s"
    )
    print(f"   Consider testing higher sigma values.")

print(f"\n{'='*100}")
```

## Step 6: Q-Q Plots (Quantile-Quantile) - Visual Gaussian Test

```python
# Q-Q plots compare quantiles of data vs theoretical Gaussian
# If points lie on diagonal line, data is Gaussian-distributed
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()

# Original field
ax = axes[0]
wind_flat_clean = wind_field_original.ravel()
wind_flat_clean = wind_flat_clean[~np.isnan(wind_flat_clean)]
sample = np.random.choice(wind_flat_clean, size=min(5000, len(wind_flat_clean)), replace=False)
stats.probplot(sample, dist="norm", plot=ax)
ax.set_title("Original Wind Field\nQ-Q Plot", fontweight="bold")
ax.grid(True, alpha=0.3)

# Noised fields
for i, sigma in enumerate(SIGMA_LEVELS):
    ax = axes[i + 1]

    noised = noised_fields[sigma]
    noised_flat = noised.ravel()
    noised_flat = noised_flat[~np.isnan(noised_flat)]

    # Sample for Q-Q plot
    sample = np.random.choice(noised_flat, size=min(5000, len(noised_flat)), replace=False)
    stats.probplot(sample, dist="norm", plot=ax)

    looks_gaussian = stats_df.iloc[i].get("looks_gaussian", False)
    snr = stats_df.iloc[i]["signal_to_noise_ratio"]

    color = "green" if looks_gaussian else "red"
    ax.set_title(
        f"σ={sigma:.1f} m/s (SNR={snr:.2f})\nGaussian: {looks_gaussian}", fontsize=10, color=color, fontweight="bold"
    )
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.suptitle("Q-Q Plots: Points on Diagonal = Gaussian Distribution", y=1.00, fontsize=14, fontweight="bold")
plt.show()

print("\nInterpretation:")
print("- Points on diagonal line → Field follows Gaussian distribution")
print("- Points deviating from line → Field retains non-Gaussian structure (cyclone signal)")
print("- Green titles indicate statistically Gaussian (p > 0.05)")
print("- Red titles indicate non-Gaussian (cyclone structure still visible)")
```

## Conclusions and Recommendations

Based on this analysis, you can:

1. **Identify Noise Threshold**: The sigma value where the cyclone structure becomes indistinguishable from pure Gaussian noise
2. **Signal-to-Noise Ratio**: Understand at what SNR the cyclone information is lost
3. **Next Steps**:
   - Use noised fields with different sigma as input to your model
   - Test if the model can recover cyclone structure from noisy observations
   - Compare model predictions with different noise levels
   - Test different sampling strategies on the noised fields

### Key Metrics:
- **SNR < 1**: Signal weaker than noise (severe degradation)
- **SNR ~ 1**: Signal comparable to noise (critical threshold)
- **SNR > 1**: Signal stronger than noise (structure preserved)
- **Shapiro-Wilk p > 0.05**: Field statistically indistinguishable from Gaussian noise

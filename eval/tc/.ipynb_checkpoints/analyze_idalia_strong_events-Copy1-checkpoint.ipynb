# TC Idalia - Strong Event Identification

This notebook identifies the strongest tropical cyclone instances from ENFO_O320 and EEFO_O96 ensemble forecasts for Hurricane Idalia.

## Objectives:
1. Load ensemble forecast data (ENFO_O320 and EEFO_O96)
2. Identify which members have the strongest cyclones (lowest MSLP, highest winds)
3. Determine which base_times and forecast steps show peak intensity
4. Extract spatial location of the cyclone centers
5. Rank and export the strongest instances for further experiments

```python
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from icecream import ic

# Add the current directory to path
sys.path.insert(0, os.getcwd())

from identify_strong_events import StrongEventIdentifier, CycloneInstance

%load_ext autoreload
%autoreload 2
```

## Configuration

```python
# Data configuration
DATA_DIR = "/home/ecm5702/hpcperm/data/tc"
EVENT_NAME = "idalia"
DATES = ["20230826", "20230827", "20230828", "20230829", "20230830"]
N_MEMBERS = 10  # Changed from 50 to 10 as requested
TOP_K = 30  # Number of strongest instances to identify
```

## Check Data Availability

```python
# Check if data files exist
event_dir = os.path.join(DATA_DIR, EVENT_NAME)
print(f"Event directory: {event_dir}")
print(f"Directory exists: {os.path.exists(event_dir)}\n")

if os.path.exists(event_dir):
    files = sorted([f for f in os.listdir(event_dir) if f.endswith(".grib")])
    print(f"Found {len(files)} GRIB files")
else:
    print("WARNING: Event directory does not exist!")
    print("You may need to run the MARS retrieval script first.")
```

## Step 1: Load and Inspect a Single GRIB File

Let's start by loading one GRIB file to understand the data structure before running the full analysis.

```python
import earthkit.data as ekd
import xarray as xr

# Load one GRIB file to inspect structure
test_date = "20230826"
test_file = f"{event_dir}/surface_pf_ENFO_O320_0001_{test_date}.grib"

print(f"Loading test file: {test_file}")
print(f"File exists: {os.path.exists(test_file)}")

if os.path.exists(test_file):
    # Load with earthkit
    print("\nLoading GRIB data...")
    data = ekd.from_source("file", test_file)
    print(f"Loaded {len(data)} GRIB messages")
else:
    print("\n⚠️ File not found! Make sure you've run the MARS retrieval script.")
```

```python
if os.path.exists(test_file):
    # Convert to xarray Dataset
    ds = data.to_xarray(engine="cfgrib")

    print("Dataset structure:")
    print("=" * 60)
    print(ds)
    print("\n" + "=" * 60)
    print(f"\nDimensions: {list(ds.dims.keys())}")
    print(f"Coordinates: {list(ds.coords.keys())}")
    print(f"Data variables: {list(ds.data_vars.keys())}")
    print(f"\nNumber of members: {len(ds.number) if 'number' in ds.dims else 'N/A'}")
    print(f"Number of forecast steps: {len(ds.step) if 'step' in ds.dims else len(ds.valid_time)}")
    print(f"Spatial dimensions: {ds.latitude.shape} x {ds.longitude.shape}")
```

### Extract and Visualize MSLP for One Member

Let's extract one member and one forecast step to visualize the MSLP field.

```python
ds["msl"]
```

```python
if os.path.exists(test_file):
    # Select one member and one forecast step
    test_member = 0
    test_step = 2  # Third forecast step

    # Extract MSLP - this will be 1D array of grid points
    mslp = ds["msl"].isel(number=test_member, step=test_step, surface=0, time=0).values
    
    # Convert to hPa if in Pa
    if np.nanmean(mslp) > 50000:
        mslp = mslp / 100

    # Get coordinate arrays (1D arrays of irregular grid points)
    lats = ds["latitude"].values
    lons = ds["longitude"].values

    print(f"Member {test_member}, Step {test_step}")
    print(f"Data shape: {mslp.shape}")
    print(f"Number of grid points: {len(mslp)}")
    print(f"MSLP range: {np.nanmin(mslp):.1f} - {np.nanmax(mslp):.1f} hPa")
    print(f"Minimum MSLP: {np.nanmin(mslp):.1f} hPa")

    # Plot using tricontourf for unstructured grid
    fig, ax = plt.subplots(1, 1, figsize=(12, 8), subplot_kw={"projection": ccrs.PlateCarree()})

    # Add map features
    ax.add_feature(cfeature.LAND, facecolor="lightgray", alpha=0.5)
    ax.add_feature(cfeature.OCEAN, facecolor="lightblue", alpha=0.3)
    ax.add_feature(cfeature.COASTLINE, linewidth=0.5)
    ax.add_feature(cfeature.BORDERS, linewidth=0.5, linestyle=":")

    # Plot MSLP using tricontourf (for unstructured grids)
    cf = ax.tricontourf(
        lons, lats, mslp, 
        levels=np.arange(980, 1020, 2), 
        cmap="RdYlBu_r", 
        extend="both", 
        transform=ccrs.PlateCarree()
    )

    # Add contour lines
    cs = ax.tricontour(
        lons, lats, mslp,
        levels=np.arange(980, 1020, 4),
        colors="black",
        linewidths=0.5,
        alpha=0.5,
        transform=ccrs.PlateCarree(),
    )
    ax.clabel(cs, inline=True, fontsize=8, fmt="%d")

    # Colorbar
    cbar = plt.colorbar(cf, ax=ax, orientation="horizontal", pad=0.05, fraction=0.046)
    cbar.set_label("Mean Sea Level Pressure (hPa)", fontsize=12)

    # Set extent
    ax.set_extent([-100, -70, 10, 40], crs=ccrs.PlateCarree())
    ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False, alpha=0.3)

    plt.title(f"ENFO_O320 - MSLP - Member {test_member} - Date: {test_date} - Step +{test_step}", fontsize=14, pad=15)
    plt.tight_layout()
    plt.show()
```

### Compute and Visualize 10m Wind Speed

```python
if os.path.exists(test_file):
    # Compute 10m wind speed from u10 and v10
    u10 = ds["u10"].isel(number=test_member, step=test_step, surface=0, time=0).values
    v10 = ds["v10"].isel(number=test_member, step=test_step, surface=0, time=0).values
    wind_speed = np.sqrt(u10**2 + v10**2)

    print(f"Wind speed range: {np.nanmin(wind_speed):.1f} - {np.nanmax(wind_speed):.1f} m/s")
    print(f"Maximum wind: {np.nanmax(wind_speed):.1f} m/s")

    # Plot
    fig, ax = plt.subplots(1, 1, figsize=(12, 8), subplot_kw={"projection": ccrs.PlateCarree()})

    # Add map features
    ax.add_feature(cfeature.LAND, facecolor="lightgray", alpha=0.5)
    ax.add_feature(cfeature.OCEAN, facecolor="lightblue", alpha=0.3)
    ax.add_feature(cfeature.COASTLINE, linewidth=0.5)
    ax.add_feature(cfeature.BORDERS, linewidth=0.5, linestyle=":")

    # Plot wind speed using tricontourf
    cf = ax.tricontourf(
        lons, lats, wind_speed, 
        levels=np.arange(0, 50, 2), 
        cmap="YlOrRd", 
        extend="max", 
        transform=ccrs.PlateCarree()
    )

    # Colorbar
    cbar = plt.colorbar(cf, ax=ax, orientation="horizontal", pad=0.05, fraction=0.046)
    cbar.set_label("10m Wind Speed (m/s)", fontsize=12)

    # Set extent
    ax.set_extent([-100, -70, 10, 40], crs=ccrs.PlateCarree())
    ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False, alpha=0.3)

    plt.title(
        f"ENFO_O320 - 10m Wind Speed - Member {test_member} - Date: {test_date} - Step +{test_step}",
        fontsize=14,
        pad=15,
    )
    plt.tight_layout()
    plt.show()
```

## Step 2: Full Analysis with StrongEventIdentifier

Now that we understand the data structure and tested the detection algorithm, let's use the `StrongEventIdentifier` class to analyze all members, dates, and forecast steps automatically.

### Test Cyclone Center Detection

Now let's test the cyclone center detection algorithm on this field.

```python
# Initialize the StrongEventIdentifier
identifier = StrongEventIdentifier(DATA_DIR, EVENT_NAME)
print(f"Initialized identifier for {EVENT_NAME}")
print(f"Event directory: {identifier.event_dir}")
```

```python
if os.path.exists(test_file):
    from scipy.ndimage import minimum_filter

    # Find cyclone center (minimum MSLP) - for 1D unstructured grid
    min_idx = np.argmin(mslp)  # Just get the index directly for 1D array
    center_lat = lats[min_idx]
    center_lon = lons[min_idx]
    min_mslp_value = mslp[min_idx]

    # Find max wind in 5° radius around center
    lat_dist = np.abs(lats - center_lat)
    lon_dist = np.abs(lons - center_lon)
    lon_dist = np.minimum(lon_dist, 360 - lon_dist)
    mask = (lat_dist < 5.0) & (lon_dist < 5.0)
    max_wind_near_center = np.nanmax(wind_speed[mask]) if np.any(mask) else np.nanmax(wind_speed)

    print(f"Cyclone Center Detection:")
    print(f"  Location: ({center_lat:.2f}°N, {center_lon:.2f}°E)")
    print(f"  Min MSLP: {min_mslp_value:.1f} hPa")
    print(f"  Max wind (5° radius): {max_wind_near_center:.1f} m/s")

    # Calculate intensity score
    mslp_normalized = (1010 - min_mslp_value) / 60
    wind_normalized = max_wind_near_center / 50
    intensity_score = (mslp_normalized + wind_normalized) / 2
    print(f"  Intensity Score: {intensity_score:.3f}")

    # Visualize with center marked
    fig, axes = plt.subplots(1, 2, figsize=(18, 8), subplot_kw={"projection": ccrs.PlateCarree()})

    for ax, field, title, cmap, levels in zip(
        axes,
        [mslp, wind_speed],
        ["MSLP (hPa)", "10m Wind Speed (m/s)"],
        ["RdYlBu_r", "YlOrRd"],
        [np.arange(980, 1020, 2), np.arange(0, 50, 2)],
    ):
        ax.add_feature(cfeature.LAND, facecolor="lightgray", alpha=0.5)
        ax.add_feature(cfeature.OCEAN, facecolor="lightblue", alpha=0.3)
        ax.add_feature(cfeature.COASTLINE, linewidth=0.5)

        # Use tricontourf for unstructured grid
        cf = ax.tricontourf(lons, lats, field, levels=levels, cmap=cmap, extend="both", transform=ccrs.PlateCarree())
        plt.colorbar(cf, ax=ax, orientation="horizontal", pad=0.05, fraction=0.046, label=title)

        # Mark cyclone center
        ax.plot(
            center_lon,
            center_lat,
            "r*",
            markersize=20,
            markeredgecolor="black",
            markeredgewidth=1.5,
            transform=ccrs.PlateCarree(),
            label="Cyclone Center",
        )

        # Draw 5° radius circle
        circle_lats = center_lat + 5 * np.cos(np.linspace(0, 2 * np.pi, 100))
        circle_lons = center_lon + 5 * np.sin(np.linspace(0, 2 * np.pi, 100))
        ax.plot(
            circle_lons,
            circle_lats,
            "r--",
            linewidth=2,
            alpha=0.7,
            transform=ccrs.PlateCarree(),
            label="5° Search Radius",
        )

        ax.set_extent([-100, -70, 10, 40], crs=ccrs.PlateCarree())
        ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False, alpha=0.3)
        ax.legend(loc="upper left")
        ax.set_title(title, fontsize=12)

    plt.suptitle(
        f"Cyclone Center Detection - Member {test_member}, Step +{test_step}\n"
        f"Center: ({center_lat:.2f}°N, {center_lon:.2f}°W) | "
        f"MSLP: {min_mslp_value:.1f} hPa | Wind: {max_wind_near_center:.1f} m/s | "
        f"Score: {intensity_score:.3f}",
        fontsize=14,
        y=1.0,
    )
    plt.tight_layout()
    plt.show()
```

### Visualize Cyclone Center Locations on Map

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

# Plot by forecast step
for member in df_enfo["member"].unique():
    df_member = df_enfo[df_enfo["member"] == member]
    axes[0].plot(df_member["forecast_step"], df_member["min_mslp"], alpha=0.3, color="blue", linewidth=1)

for member in df_eefo["member"].unique():
    df_member = df_eefo[df_eefo["member"] == member]
    axes[0].plot(df_member["forecast_step"], df_member["min_mslp"], alpha=0.3, color="orange", linewidth=1)

axes[0].set_xlabel("Forecast Step (hours)")
axes[0].set_ylabel("Minimum MSLP (hPa)")
axes[0].set_title("MSLP Evolution by Forecast Step")
axes[0].grid(True, alpha=0.3)
axes[0].invert_yaxis()  # Lower MSLP is stronger

# Plot by base time
df_enfo_grouped = df_enfo.groupby("base_time")["min_mslp"].agg(["mean", "min", "max"])
df_eefo_grouped = df_eefo.groupby("base_time")["min_mslp"].agg(["mean", "min", "max"])

axes[1].plot(
    df_enfo_grouped.index, df_enfo_grouped["mean"], marker="o", label="ENFO_O320 mean", color="blue", linewidth=2
)
axes[1].fill_between(df_enfo_grouped.index, df_enfo_grouped["min"], df_enfo_grouped["max"], alpha=0.2, color="blue")

axes[1].plot(
    df_eefo_grouped.index, df_eefo_grouped["mean"], marker="^", label="EEFO_O96 mean", color="orange", linewidth=2
)
axes[1].fill_between(df_eefo_grouped.index, df_eefo_grouped["min"], df_eefo_grouped["max"], alpha=0.2, color="orange")

axes[1].set_xlabel("Base Time")
axes[1].set_ylabel("Minimum MSLP (hPa)")
axes[1].set_title("MSLP Evolution by Base Time (shaded area shows min-max range)")
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].invert_yaxis()
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("idalia_temporal_evolution.png", dpi=150, bbox_inches="tight")
plt.show()
```

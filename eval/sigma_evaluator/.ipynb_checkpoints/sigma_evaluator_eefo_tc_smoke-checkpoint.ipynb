# EEFO O96 TC Sigma Evaluator CPU Smoke Test

This notebook runs a minimal sigma-evaluator smoke test on CPU (no GPU required), with both:
- noised target (`prediction_on_pure_noise=False`)
- pure-noise target (`prediction_on_pure_noise=True`)

It also anchors the run context on the large EEFO O96 Idalia GRIB files.

```python
import os

os.environ["DEV"] = "/home/ecm5702/dev/"
os.environ["DATA_DIR"] = "/home/mlx/ai-ml/datasets/"
os.environ["DATA_STABLE_DIR"] = "/home/mlx/ai-ml/datasets/stable/"
os.environ["OUTPUT"] = "/ec/res4/scratch/ecm5702/aifs"
os.environ["GRID_DIR"] = "/home/mlx/ai-ml/grids/"
os.environ["INTER_MAT_DIR"] = "/ec/res4/hpcperm/nesl/inter_mat/"
os.environ["RESIDUAL_STATISTICS_DIR"] = (
    "/home/ecm5702/hpcperm/data/residuals_statistics/"
)
os.environ["GRAPHS"] = "/home/ecm5702/scratch/aifs/graphs"
os.environ["SLURM_GPUS_PER_NODE"] = "1"
os.environ["SLURM_NNODES"] = "1"
```

```python
import os
import glob
import subprocess
from datetime import datetime
import pandas as pd

DOWNTOOLS = "/home/ecm5702/dev/downscaling-tools"
SIGMA_DIR = f"{DOWNTOOLS}/eval_suite/sigma_evaluator"
CKPT_DIR = "/home/ecm5702/scratch/aifs/checkpoint"
TC_DIR = "/home/ecm5702/hpcperm/data/tc/idalia"

NAME_EXP = "0e1602b36722471f97a3dfa93b81297e"
NAME_CKPT = "anemoi-by_time-epoch_021-step_100000.ckpt"
N_SAMPLES = 1
SIGMAS = "0.5"
VALID_FREQ = "50h"

run_tag = datetime.now().strftime("%Y%m%d_%H%M%S")
OUT_FILE = f"sigma_eval_smoke_cpu_eefo_{run_tag}.csv"
OUT_CSV = f"{CKPT_DIR}/{NAME_EXP}/{OUT_FILE}"

print('Sigma evaluator dir:', SIGMA_DIR)
print('Output CSV:', OUT_CSV)
```

```python
# Pick one of the largest EEFO_O96 files to anchor the TC test context
eefo_files = sorted(glob.glob(f"{TC_DIR}/surface_pf_EEFO_O96_0001_*.grib"))
file_sizes = sorted([(os.path.getsize(fp), fp) for fp in eefo_files], reverse=True)

print('EEFO_O96 files by size (top 6):')
for sz, fp in file_sizes[:6]:
    print(f"  {sz/1024/1024:.2f} MB  {os.path.basename(fp)}")

largest_file = file_sizes[0][1] if file_sizes else None
print('\nSelected largest file:', largest_file)
```

```python
# Run sigma evaluator on CPU in both modes (noised + pure-noise)
cmd = [
    'python', '-u', f'{SIGMA_DIR}/run_sigma_evaluator.py',
    '--name_exp', NAME_EXP,
    '--name_ckpt', NAME_CKPT,
    '--out_file', OUT_FILE,
    '--device', 'cpu',
    '--n_samples', str(N_SAMPLES),
    '--sigmas', SIGMAS,
    '--validation_frequency', VALID_FREQ,
    '--run_noised',
    '--run_pure_noise',
]

print('Running command:')
print(' '.join(cmd))

proc = subprocess.run(
    cmd,
    cwd=SIGMA_DIR,
    text=True,
    capture_output=True,
)

print('\nReturn code:', proc.returncode)
print('\n--- STDOUT (last 4000 chars) ---')
print(proc.stdout[-4000:])
print('\n--- STDERR (last 4000 chars) ---')
print(proc.stderr[-4000:])
```

```python
# Load and inspect resulting table if run succeeded
if os.path.exists(OUT_CSV):
    df = pd.read_csv(OUT_CSV)
    print('Rows:', len(df))
    display(df)
else:
    print('Output CSV not found:', OUT_CSV)
```

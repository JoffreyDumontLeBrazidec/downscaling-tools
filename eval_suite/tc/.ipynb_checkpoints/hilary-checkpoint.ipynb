```python
import cartopy.crs as crs  # can't be called after metview for some reason


# load modules


import matplotlib.pyplot as plt
import xarray as xr


import numpy as np
from time import sleep
import os as os
import glob as glob
import netCDF4 as nc
import seaborn as sns
from icecream import ic
import warnings
import seaborn as sns

warnings.filterwarnings(
    "ignore",
    message=".*decode_timedelta will default to False.*",
    category=FutureWarning,
    module="cfgrib.xarray_plugin",
)


sns.set_theme(style="ticks", rc={"font.family": "DejaVu Sans"})

# select subdomain and resolution
year = "2023"
month = "08"


dates = ["16", "17", "18", "19", "20"]


list_expid_ml = [
    "ENFO_O320_ip6y",
    "ENFO_O320_iz2q",
    "ENFO_O320_iytb",
    "ENFO_O320_iyl7",
    "ENFO_O320_iz2r",
    "ENFO_O320_iz2s",
    "ENFO_O320_iyte",
    "ENFO_O320_iyjp",
]
expid_enfo_O320 = "ENFO_O320_0001"
expid_enfo_O96 = "ENFO_O96_0001"
expid_eefo_O320 = "EEFO_O320_0001"
expid_eefo_O96 = "EEFO_O96_0001"
analysis = "OPER_O320_0001"

list_exps = [
    analysis,
    expid_enfo_O320,
    expid_enfo_O96,
    expid_eefo_O320,
    expid_eefo_O96,
] + list_expid_ml

main_dir = "/home/ecm5702/dev/post_prepml/tc"
```

```python
import earthkit.data as ekd
```

```python
event = "hilary"
dir_data = f"/home/ecm5702/hpcperm/data/tc/{event}"
```

```python
all_datasets = {}
for exp in list_expid_ml:
    ic(exp)
    files = [
        os.path.join(dir_data, f"surface_pf_{exp}_{year}{month}{day}.grib")
        for day in dates
    ]
    data = ekd.from_source("file", files)
    ds = data.to_xarray(engine="cfgrib")
    ds = ds.assign_coords(exp=exp)
    all_datasets[exp] = ds
```

```python
files = [
    os.path.join(dir_data, f"surface_pf_ENFO_O320_0001_{year}{month}{day}.grib")
    for day in dates
]
dataset_enfo_o320 = ekd.from_source("file", files).to_xarray(engine="cfgrib")
```

```python
files = [
    os.path.join(dir_data, f"surface_pf_EEFO_O96_0001_{year}{month}{day}.grib")
    for day in dates
]
dataset_eefo_o96 = ekd.from_source("file", files).to_xarray(engine="cfgrib")
```

```python
dataset_oper = ekd.from_source(
    "file",
    [
        os.path.join(dir_data, f"surface_an_OPER_O320_0001_20230816.grib"),
    ],
).to_xarray(engine="cfgrib")
```

```python
sns.set_theme(style="ticks", rc={"font.family": "DejaVu Sans"})
```

```python
import cmcrameri.cm as cm

cmap = cm.batlow  # or any other, e.g. cm.roma, cm.lapaz
colors = cmap(np.linspace(0, 1, len(all_datasets)))
```

```python
import numpy as np

np.seterr(divide="ignore", invalid="ignore")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
xbins_msl = np.arange(960, 1021, 1)
bin_mids = (xbins_msl[:-1] + xbins_msl[1:]) / 2

hist_oper, _ = np.histogram(
    dataset_oper["msl"].values.flatten() / 100, bins=xbins_msl, density=True
)

for i, exp in enumerate(all_datasets.keys()):
    vals = all_datasets[exp]["msl"].values.flatten() / 100
    hist_exp, _ = np.histogram(vals, bins=xbins_msl, density=True)
    axs[0].plot(
        bin_mids,
        hist_exp / hist_oper,
        linestyle="-",
        linewidth=3,
        label=exp,
        color=colors[i],
    )

vals_enfo = dataset_enfo_o320["msl"].values.flatten() / 100
hist_enfo_o320, _ = np.histogram(vals_enfo, bins=xbins_msl, density=True)
axs[0].plot(
    bin_mids,
    hist_enfo_o320 / hist_oper,
    linestyle="-.",
    linewidth=2,
    color="black",
    label="enfo_o320",
)

vals_eefo = dataset_eefo_o96["msl"].values.flatten() / 100
hist_eefo_o96, _ = np.histogram(vals_eefo, bins=xbins_msl, density=True)
axs[0].plot(
    bin_mids,
    hist_eefo_o96 / hist_oper,
    linestyle="--",
    linewidth=2,
    color="red",
    label="eefo_o96",
)

axs[0].plot(
    bin_mids,
    np.ones_like(bin_mids),
    linestyle="--",
    linewidth=2,
    color="green",
    label="OPER",
)
axs[0].set_ylim(0, 2)
axs[0].legend()
axs[0].set_xlabel("Mean Sea Level Pressure (hPa)", fontsize=14)
axs[0].set_ylabel("Normalized Probability Density", fontsize=14)
axs[0].set_title("Normalized (by AN O320) Distribution MSLP", fontsize=14)

# second plot, e.g. (u10+v10)**2
xbins_wind10m = np.arange(0, 30.01, 1)
bin_mids = (xbins_wind10m[:-1] + xbins_wind10m[1:]) / 2

hist_oper, _ = np.histogram(
    np.sqrt(dataset_oper["u10"] ** 2 + dataset_oper["v10"] ** 2).values.flatten(),
    bins=xbins_wind10m,
    density=True,
)

for i, exp in enumerate(all_datasets.keys()):
    vals = np.sqrt(
        all_datasets[exp]["u10"] ** 2 + all_datasets[exp]["v10"] ** 2
    ).values.flatten()
    hist_exp, _ = np.histogram(vals, bins=xbins_wind10m, density=True)
    axs[1].plot(
        bin_mids,
        hist_exp / hist_oper,
        linestyle="-",
        linewidth=3,
        label=exp,
        color=colors[i],
    )

vals_enfo = np.sqrt(
    dataset_enfo_o320["u10"] ** 2 + dataset_enfo_o320["v10"] ** 2
).values.flatten()
hist_enfo_o320, _ = np.histogram(vals_enfo, bins=xbins_wind10m, density=True)
axs[1].plot(
    bin_mids,
    hist_enfo_o320 / hist_oper,
    linestyle="-.",
    linewidth=2,
    color="black",
    label="enfo_o320",
)

vals_eefo = np.sqrt(
    dataset_eefo_o96["u10"] ** 2 + dataset_eefo_o96["v10"] ** 2
).values.flatten()
hist_eefo_o96, _ = np.histogram(vals_eefo, bins=xbins_wind10m, density=True)
axs[1].plot(
    bin_mids,
    hist_eefo_o96 / hist_oper,
    linestyle="--",
    linewidth=2,
    color="red",
    label="eefo_o96",
)

axs[1].plot(
    bin_mids,
    np.ones_like(bin_mids),
    linestyle="--",
    linewidth=2,
    color="green",
    label="OPER AN",
)
axs[1].set_ylim(0, 2)
axs[1].set_xlabel("10m wind speed (m/s)", fontsize=14)
axs[1].set_ylabel("Normalized Probability Density", fontsize=14)
axs[1].set_title("Normalized (by AN O320) Distribution 10m Wind Speed", fontsize=14)
fig.suptitle("Hilary normed pdfs")
plt.tight_layout()

# filename = "hilary_norm_pdfs.pdf"
# path_fig = os.path.join("/home/ecm5702/dev/outputs/tc", filename)
# ic(f"Saving figure to {path_fig}")
# plt.savefig(path_fig, dpi=300, bbox_inches="tight")
```
